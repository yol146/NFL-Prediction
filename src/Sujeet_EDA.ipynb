```python
df = df.loc[~df["Tm_QBRating"].isnull()]
```

```python
x_cols = ["Tm_QBRating", "Opp_QBRating", "Home"]

import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler

sf = StandardScaler()
sf.fit(df[x_cols])

X = sm.add_constant(sf.transform(df[x_cols]))
y = df.Spread

model = sm.OLS(y, X)
results = model.fit()

results.params
```
